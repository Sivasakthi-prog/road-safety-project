```python
import matplotlib.pyplot as plt

# Data for the bar chart
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

# Create the bar chart
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='purple')

# Add labels and title
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Select only numeric columns for correlation calculation
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
corr_matrix = numeric_tips.corr()

# Create the heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Add a title
plt.title("Correlation Heatmap")

# Display the plot
plt.show()

```

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Create the DataFrame
df = pd.DataFrame({
    "x": np.linspace(0, 10, 100),
    "y": np.sin(np.linspace(0, 10, 100))
})

# Create the plot
fig = px.line(df, x='x', y='y', title="Interactive Sine Wave")

# Show the plot
fig.show()
```

```python
import seaborn as sns
import plotly.express as px

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create the scatter plot
fig = px.scatter(tips, x='total_bill', y='tip', color='sex', size='size', title="Total Bill vs Tip")

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go
import seaborn as sns

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create the 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=tips['total_bill'],
    y=tips['tip'],
    z=tips['size'],
    mode='markers',
    marker=dict(size=5, color=tips['total_bill'], colorscale='Viridis')
)])

# Update layout with a title
fig.update_layout(title="3D Scatter Plot of Total Bill, Tip & Size")

# Show the plot
fig.show()
```

```python
# Assuming df is already defined, and columns exist in the dataset
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Annual_Income"].fillna(df["Annual_Income"].median(), inplace=True)
df["Spending_Score"].fillna(df["Spending_Score"].mean(), inplace=True)
df["Purchase_Frequency"].fillna(df["Purchase_Frequency"].median(), inplace=True)

# Correcting the 'True' instead of 'TRUE' in the 'Last_Transaction_Days' column
df["Last_Transaction_Days"].fillna(df["Last_Transaction_Days"].mean(), inplace=True)

# Fixing 'Preferred_Category' to fill NaN with the mode, instead of the median
df["Preferred_Category"].fillna(df["Preferred_Category"].mode()[0], inplace=True)

print(df)
```
