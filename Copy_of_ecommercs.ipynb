```python
import pandas as pd

# Use read_excel instead of read_xlsx
df = pd.read_excel("ecommerce_datafrem.xlsx")
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df_cleaned=df.dropna()
print(df_cleaned)
```

```python
from pickle import TRUE
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Annual_Income"].fillna(df["Annual_Income"].median(), inplace=True)
df["Spending_Score"].fillna(df["Spending_Score"].mean(), inplace=True)
df["Purchase_Frequency"].fillna(df["Purchase_Frequency"].median(), inplace=TRUE)
df["Last_Transaction_Days"].fillna(df["Last_Transaction_Days"].mean(), inplace=TRUE)
df["Preferred_Category"].fillna(df["Preferred_Category"].median(), inplace=TRUE)
print(df)
```

```python
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Annual_Income"].fillna(df["Annual_Income"].mean(), inplace=True)
df["Spending_Score"].fillna(df["Spending_Score"].mean(), inplace=True)
df["Purchase_Frequency"].fillna(df["Purchase_Frequency"].median(), inplace=True)
df["Last_Transaction_Days"].fillna(df["Last_Transaction_Days"].mean(), inplace=True)
print(df)
```
